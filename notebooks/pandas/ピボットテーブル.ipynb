```python
import pandas as pd

# データフレームの作成
data = {"Name":["Alice","Bob","Charlie","David","Eve","Hiroko"],
        "Age":[25,30,35,40,45,31],}

df = pd.DataFrame(data)
df
```

```python
df["Category"] = ["A","B","C","A","B","A"]
df["Gender"] = ["F","M","F","M","F","F"]
```

```python
df
```

```python
# ピボットテーブルとは、データを集計するためのテーブル
df.pivot_table(index="Category",columns="Gender",values="Age",aggfunc="sum")
```

```python
df.pivot_table(index="Gender",columns="Category",values="Age",aggfunc="mean")
```
