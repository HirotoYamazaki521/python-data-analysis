```python
import pandas as pd

data = {"Name": ['Eve', 'Frank', 'Grace', 'Hank'],
        "Date": ['2021-06-15', '2022-08-10', '2023-01-20', '2020-03-25'],
        "Salary": [55000, 62000, 58000, 70000]}

df = pd.DataFrame(data)
df
```

```python
# 列 Date を日付データ型に変換する。
df["Date"] = pd.to_datetime(df["Date"])
df
```

```python
# 列 Year を新たに作成し、Date の年を抽出して格納する。
df["Year"] = df["Date"].dt.year
df
```

```python
# 列 Salary を降順にソートする
df.sort_values(by="Salary",ascending=False)
```
