```python
import pandas as pd

# データフレームの作成
data = {"Name":["Alice","Bob","Charlie","David","Eve"],
        "Age":[25,30,35,40,45],}

df = pd.DataFrame(data)
df
```

```python
df2 = pd.DataFrame({"Name":["Alice","Bob"],"Salary":[50000,60000]})
df2
```

```python
# 縦方向に結合
pd.concat([df,df2])
```

```python
# axis=1 で横方向に結合
pd.concat([df,df2],axis=1)
```
