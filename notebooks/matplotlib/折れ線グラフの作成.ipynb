```python
import matplotlib.pyplot as plt

# やってることは一緒だけど
# from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 折れ線グラフを作成
plt.plot(x, y)

# グラフのタイトルを設定
plt.title("Line Graph Example")

# x軸のラベルを設定
plt.xlabel("X-axis")

# y軸のラベルを設定
plt.ylabel("Y-axis")

# グラフを表示
plt.show()
```
