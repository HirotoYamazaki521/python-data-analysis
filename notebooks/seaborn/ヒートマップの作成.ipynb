```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
 
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [500, 600, 750, 850, 900],
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture'],
    'Ad_Spend': [100, 200, 300, 400, 500],
    'Values': [7, 8, 9, 5, 4],
    'Region': ['North', 'South', 'East', 'West', 'North']
}
 
df = pd.DataFrame(data)
df
```

```python
# ヒートマップの作成
# 豆知識
# データ間の関係が強いことを相関関係が強いという
# 相関が強いと言うのは相関係数が1とか-1とかに近い値になる
corr =df[["Sales","Ad_Spend","Values"]].corr()

# annotは相関係数を表示する
# cmapはカラーマップを指定する
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```
