```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
 
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [500, 600, 750, 850, 900],
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture'],
    'Ad_Spend': [100, 200, 300, 400, 500],
    'Values': [7, 8, 9, 5, 4],
    'Region': ['North', 'South', 'East', 'West', 'North']
}
 
df = pd.DataFrame(data)
df
```

```python
# ペアプロットの作成
# ペアプロットとは、2つの変数の関係を可視化するための方法
# 例：商品の売上と広告費用の関係をペアプロットで可視化する
sns.pairplot(df[["Sales","Ad_Spend","Values"]])
plt.show()
```
